# Games of Chance
Programming done in Python

#### Importing library for random number generation

```python
import random
```

```python
# Flipping a Coin Function

def coin_flip(guess, bet):
    
    # Bet must be larger than 0
    if bet <= 0:
        print("--------------------")
        print("Your bet should be above 0.")
        print("--------------------")
        return 0
    
    # Starts the game and flips a coin
    print("--------------------")
    print("Let's flip a coin! You guessed " + guess + ".")
    num = random.randint(1, 2)
    
    # Results of the coin flip
    if num == 1:
        print("Heads")
    else:
        print("Tails")
        
    # If player wins or loses the bet
    if (guess == "Heads" and num == 1) or (guess == "Tails" and num == 2):
        print("You won " + str(bet) + " dollars!")
        print("--------------------")
        return bet
    else:
        print("You lost " + str(bet) + " dollars!")
        print("--------------------")
        return -bet
```

```python
# Cho-Han Function

def cho_han(guess, bet):
    
    # Bet must be larger than 0
    if bet <= 0:
        print("--------------------")
        print("Your bet should be above 0.")
        print("--------------------")
        return 0
    
    # Game starts by rolling the dice and getting players guesses and bets
    dice_1 = random.randint(1, 6)
    dice_2 = random.randint(1, 6)
    total = dice_1 + dice_2
    print("--------------------")
    print("Is the total Odd or Even? Player guesses " + guess + " and bets " + str(bet) + ' dollars.')
    print('Total = ' + str(total))
    
  
    # Total of dice and if player won or lost
    if (guess == 'Even' and total % 2 == 0) or (guess == 'Odd' and total % 2 == 1):
        print('You won ' + str(bet) + ' dollars!')
        print('--------------------')
        return bet
    else:
        print('You lost ' + str(bet) + ' dollars!')
        print('--------------------')
        return -bet
```

```python
# Card Drawing Function

def card_draw(bet):
  
  # Bet must be larger than 0
  if bet <= 0:
    print('--------------------')
    print('Your bet should be above 0.')
    print('--------------------')
    return 0
  
  # Drawing cards from the deck
  p1 = random.randint(1, 13)
  p2 = random.randint(1, 13)
  print('--------------------')
  print("Let's draw cards. The player with the highest card wins.")
  print('Your Card = ' + str(p1))
  print('Their Card = ' + str(p2))
  
  # Results if win/lose or tie
  if p1 > p2:
    print('You won ' + str(bet) + ' dollars!')
    print('--------------------')
    return bet
  elif p1 < p2:
    print('You lost ' + str(bet) + ' dollars!')
    print('--------------------')
    return -bet
  else:
    print('Tie! Would you like to play again?')
    print('--------------------')
    return 0
```

```python
# Roulette Function

def roulette(guess, bet):
  
  # Bet must be larger than 0.
  if bet <= 0:
    print('--------------------')
    print('Your bet should be above 0.')
    print('--------------------')
    return 0
  
  # Asking for player guess and bet
  print('--------------------')
  print("Let's play roulette!")
  result = random.randint(0, 37)
  
  # If result is 37 we count this as 00
  if result == 37:
    print('The ball landed on 00')
  else:
    print('The ball landed on ' + str(result))
    
  # If ball lands on Even and not 0 and the guess was even
  if guess == 'Even' and  result % 2 == 0 and result != 0:
    print(str(result) + ' is an Even number.')
    print('You won ' + str(bet) + ' dollars!')
    print('--------------------')
    return bet
  
  # If ball lands on Odd and not 00 and the guess was odd
  elif guess == 'Odd' and result % 2 == 1 and result != 37:
    print(str(result) + ' is an Odd number.')
    print('You won ' + str(bet) + ' dollars!')
    print('--------------------')
    return bet
  
  # If player bets a number and ball lands on that number, player wins 35 times their bet
  elif guess == result:
    print('You guessed ' + str(guess) + ' and the ball landed on ' + str(result) + '.')
    print('You won ' + str(bet*35) + ' dollars!')
    print('--------------------')
    return bet * 35
  
  else:
    print('You lost ' + str(bet) + ' dollars!')
    print('--------------------')
    return -bet
```
